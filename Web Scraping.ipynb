```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_html_content(url):
    response = requests.get(url)
    return BeautifulSoup(response.content, 'html.parser')
```

```python
def extract_article_text_with_headings(soup):
    article_data = {}
    headings = soup.find_all(['h2', 'h3', 'h4', 'h5', 'h6'])
    
    for heading in headings:
        heading_text = heading.get_text()
        content = [p.get_text() for p in heading.find_next_siblings('p')]
        article_data[heading_text] = content
    
    return article_data

print(scraped_data['article_text_with_headings'])
```

```python
def collect_redirect_links(soup):
    links = [link['href'] for link in soup.select('a[href^="/wiki/"]:not([href^="/wiki/File:"]):not([href^="/wiki/Category:"])')]
    return links
```

```python
def scrape_wikipedia_page(url):
    soup = get_html_content(url)
    title = extract_article_title(soup)
    article_text_with_headings = extract_article_text_with_headings(soup)
    redirect_links = collect_redirect_links(soup)
    
    scraped_data = {
        'title': title,
        'article_text_with_headings': article_text_with_headings,
        'redirect_links': redirect_links
    }
    
    return scraped_data

wikipedia_url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'
scraped_data = scrape_wikipedia_page(wikipedia_url)

print(scraped_data['redirect_links'])
```
